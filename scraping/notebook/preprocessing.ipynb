```python
import os
import pickle
import re

import pandas as pd
pd.set_option("max_rows", 500, "max_colwidth", 500, "max_columns", 500)
import numpy as np

import MeCab
import mojimoji
```

```python
with open("./article_infos.bin", "rb") as f:
    artilcle_infos = pickle.load(f)
```

```python
artilcle_infos[0]
```

```python
df = pd.DataFrame(artilcle_infos, columns=["url", "date", "title", "context", "category", "keyword", "tweet_share", "quote"])
df.head()
```

```python
print(df.shape)
df = df.query("title != 'None Title'")
print(df.shape)
```

```python
# date前処理
df["date"] = pd.to_datetime(df.date, format="%Y年%m月%d日 %H時%M分")
df.date.head()
```

```python
# tweet_share前処理
df = df.astype({"tweet_share": np.int32})
df.info()
```

```python
def exclude_main_category(x):
    try:
        return x.split("<SEP>")[1]
    except:
        return "None"

df["main_category"] = df.category.apply(exclude_main_category)
df.head()
```

```python
df.groupby("main_category").describe()
```

```python
df.query("tweet_share == 18513")
```

```python
main_category_count_df = df.groupby("main_category").url.count()
use_main_category = main_category_count_df.index[main_category_count_df > 100].to_list()
use_main_category
```

```python
print(df.shape)
df = df[df.main_category.isin(use_main_category)]
print(df.shape)
df.main_category.unique()
```

```python
# 無駄なスペース除去 + 全て全角文字に
def preprocess_context(x):
    t = re.sub(r"\s", "", x)
    # 数字を半角に
    t = mojimoji.zen_to_han(t, kana=False, ascii=False)
    # 数字以外を全角に
    t = mojimoji.han_to_zen(t, digit=False)
    return t

df["context_"] = df.context.apply(preprocess_context)
df[["context", "context_"]].head()
```
